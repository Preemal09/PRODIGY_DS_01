```python
#loading the dataset to pandas dataframe
dataset=pd.read_csv("P:/Prodigy internship/tsk1.csv",encoding="unicode_escape")
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.sample(5)
```

```python
#droping the columns which are not necessary
dataset.drop(['SpecialNotes','TableName'],axis=1,inplace=True)
```

```python
dataset
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
#number of missing values in each column
dataset.isnull().sum()
```

```python
#droping the null values from listed columns
columns=['Region','IncomeGroup']
dataset=dataset.dropna(subset=columns)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
#grouping the data based on region and income group
group_data=dataset.groupby(['Region','IncomeGroup']).size()
group_data
```

```python
#grouping the data based on region and income group using unstack()
group_data1=dataset.groupby(['Region','IncomeGroup']).size().unstack(fill_value=0)
group_data1
```

```python
#barchart representation of grouped data
group_data1.plot(kind='bar',figsize=(10,8))
plt.xlabel('Region')
plt.ylabel('Number of countries')
plt.title('Income distribution in various regions')
plt.legend(title='Income group')
```

```python
#stacked barchart representation of grouped data
group_data1.plot(kind='bar',stacked=True,figsize=(10,8))
plt.xlabel('Region')
plt.ylabel('Number of countries')
plt.title('Income distribution in various regions')
plt.legend(title='Income group')
```

```python
#countplot to represent the number of countries in each region
plt.figure(figsize=(20,20))
sns.countplot(x='Region',data=dataset)
plt.title("number of countries in each region")
plt.show()
```
